```python
import whisper

model = whisper.load_model("base")
```

```python
from IPython.display import Audio
Audio("/content/f_m.wav")
```

```python
audio = whisper.load_audio("/content/f_m.wav")
audio = whisper.pad_or_trim(audio)

mel = whisper.log_mel_spectrogram(audio).to(model.device)

_, probs = model.detect_language(mel)
print(f"Detected language: {max(probs, key=probs.get)}")

options = whisper.DecodingOptions()
result = whisper.decode(model, mel, options)

print(result.text)
```

```python
import os
from langchain_google_genai  import ChatGoogleGenerativeAI
from langchain.text_splitter import CharacterTextSplitter

os.environ['GOOGLE_API_KEY'] = ''
api_key = os.getenv('GOOGLE_API_KEY')

llm = ChatGoogleGenerativeAI(
    model="gemini-pro", temperature=0.7
)

splitter = CharacterTextSplitter(chunk_size=2000)


def summarize_text(text):
  chunks = splitter.split_text(text)
  summary = ""

  for chunk in chunks:

    response = llm.invoke(input=chunk)
    summary += response.content.strip()

  return summary

summary = summarize_text(result.text)
print(f"Original Text:\n{result.text}\n")
print(f"Summary:\n{summary}")
```
